```python
# Importando as bibliotecas essenciais
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurando o estilo dos gráficos para ficar com aspecto profissional
sns.set_theme(style="whitegrid")
```

```python
# Carregando o dataset oficial do desafio
url = "https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv"
print("Baixando dados... aguarde.")
df = pd.read_csv(url)

# Mostrando as 5 primeiras linhas para entender a estrutura
display(df.head())
```

```python
# Contando as classes (0 = Normal, 1 = Fraude)
contagem_classes = df['Class'].value_counts()
print("Distribuição das Classes:")
print(contagem_classes)

# Plotando um gráfico de barras em escala logarítmica (para conseguirmos enxergar a barra das fraudes)
plt.figure(figsize=(8, 5))
sns.barplot(x=contagem_classes.index, y=contagem_classes.values, hue=contagem_classes.index, palette=['#1f77b4', '#d62728'], legend=False)
plt.title('Distribuição de Transações: Normais vs Fraudes (Escala Logarítmica)')
plt.xlabel('Classe (0: Normal, 1: Fraude)')
plt.ylabel('Número de Transações')
plt.yscale('log') # Escala logarítmica é crucial aqui!
plt.show()
```

```python
# Comparando o valor das transações (Amount) entre normais e fraudes
plt.figure(figsize=(10, 6))
sns.boxplot(x='Class', y='Amount', data=df, palette=['#1f77b4', '#d62728'])
plt.title('Distribuição do Valor da Transação por Classe')
plt.xlabel('Classe (0: Normal, 1: Fraude)')
plt.ylabel('Valor da Transação (Amount)')

# Limitando o eixo Y para remover valores absurdamente extremos (outliers naturais) e facilitar a visualização
plt.ylim(-20, 400)
plt.show()
```
